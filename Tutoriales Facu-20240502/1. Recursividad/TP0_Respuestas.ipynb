## Ejercicio 1) Torre de Hanoi

### a)

```python
def hanoi(n, i, j): # n discos deben ser movidos de la vara i a la vara j (3 varas)
    solution = []
    if n == 1:
        solution.extend([(n, i, j)])
    else:
        k = 6 - i - j # 1 + 2 + 3 = 6
        solution.extend(hanoi(n-1, i, k))
        solution.extend([(n, i, j)])
        solution.extend(hanoi(n-1, k, j))
    return solution
```

\begin{align*}
&\text{Demostracin por induccin del problema de la Torre de Hanoi} \\

&\text{Paso base:} \\
&\text{Para } n = 1, \text{ el nmero mnimo de movimientos es } 1, \text{ ya que solo se tiene un disco y se puede mover directamente al poste objetivo.} \\
&S_1 = 2^1 - 1 = 1 \\
&\text{El paso base se cumple.} \\

&\text{Paso inductivo:} \\
&\text{Supongamos que la frmula es verdadera para } n = k, \text{ es decir, que se necesitan } S_k = 2^k - 1 \text{ movimientos para resolver el problema con } k \text{ discos.} \\

&\text{Para resolver el problema con } k + 1 \text{ discos, seguimos los siguientes pasos:} \\
&1. \text{Mover los primeros } k \text{ discos al poste intermedio utilizando } 2^k - 1 \text{ movimientos (por hiptesis inductiva).} \\
&2. \text{Mover el disco } k + 1 \text{ al poste objetivo, lo cual toma } 1 \text{ movimiento.} \\
&3. \text{Mover los } k \text{ discos desde el poste intermedio al poste objetivo encima del disco } k + 1, \text{ utilizando otros } 2^k - 1 \text{ movimientos.} \\

&\text{Por lo tanto, el total de movimientos necesarios es:} \\
&(2^k - 1) + 1 + (2^k - 1) = 2 \cdot 2^k - 1 = 2^{k+1} - 1 \\

&\text{Esto completa el paso inductivo y demuestra que si la frmula es verdadera para } n = k, \text{ entonces tambin es verdadera para } n = k + 1. \\

&\text{Conclusin:} \\
&\text{Por induccin matemtica, la frmula } S_n = 2^n - 1 \text{ es verdadera para cualquier nmero natural de } n, \text{ y por lo tanto, el monje tibetano est en lo cierto.}
\end{align*}

```python
import math

def Fib2(n):
    sqrt_5 = math.sqrt(5)
    phi = (1 + sqrt_5) / 2
    psi = (1 - sqrt_5) / 2
    return int((phi**n - psi**n) / sqrt_5)

fib_table = [{"n": n, "Fib1": Fib1(n), "Fib2": Fib2(n)} for n in range(1, 21)]

fib_table
```

\begin{align*}
&\textbf{La frmula de Binet para la secuencia de Fibonacci es:} \\
&F(n) = \frac{\varphi^n - \psi^n}{\sqrt{5}} \text{ donde:} \\
&\varphi = \frac{1 + \sqrt{5}}{2}, \quad \text{la proporcin urea, y} \\
&\psi = \frac{1 - \sqrt{5}}{2} = 1 - \varphi, \quad \text{el conjugado de } \varphi. \\
\\
&\textbf{Base de Induccin:} \\
&\text{Para } n = 0, \quad F(0) = 0 = \frac{\varphi^0 - \psi^0}{\sqrt{5}} = \frac{1 - 1}{\sqrt{5}} = 0. \\
&\text{Para } n = 1, \quad F(1) = 1 = \frac{\varphi^1 - \psi^1}{\sqrt{5}} = \frac{\varphi - \psi}{\sqrt{5}} = \frac{\sqrt{5}}{\sqrt{5}} = 1. \\
\\
&\textbf{Paso Inductivo:} \\
&\text{Suponemos que la frmula es vlida para } F(n) \text{ y } F(n-1), \text{ es decir:} \\
&F(n) = \frac{\varphi^n - \psi^n}{\sqrt{5}}, \\
&F(n-1) = \frac{\varphi^{n-1} - \psi^{n-1}}{\sqrt{5}}. \\
\\

&\textbf{Necesitamos probar que la frmula es vlida para } F(n+1): \\
&F(n+1) = F(n) + F(n-1) \\
&= \frac{\varphi^n - \psi^n}{\sqrt{5}} + \frac{\varphi^{n-1} - \psi^{n-1}}{\sqrt{5}} \\
&= \frac{\varphi^n + \varphi^{n-1} - (\psi^n + \psi^{n-1})}{\sqrt{5}}. \\
\\
&\textbf{Usando las propiedades de } \varphi \text{ y } \psi, \varphi^2 = \varphi + 1 \text{ y } \psi^2 = \psi + 1, \text{ podemos escribir:} \\
&\varphi^{n-1} = \frac{\varphi^n}{\varphi}, \\
&\psi^{n-1} = \frac{\psi^n}{\psi}. \\

&\text{Entonces,} \\
&F(n+1) = \frac{\varphi^n(1 + 1/\varphi) - \psi^n(1 + 1/\psi)}{\sqrt{5}} \\
&= \frac{\varphi^n(\varphi) - \psi^n(\psi)}{\sqrt{5}} \\
&= \frac{\varphi^{n+1} - \psi^{n+1}}{\sqrt{5}}. \\

&\text{Esto completa la demostracin por induccin, y por lo tanto, la frmula de Binet es vlida para todos los nmeros enteros no negativos } n.
\end{align*}

```python
def Fib3(n):
    if n <= 2:
        return 1
    f1, f2 = 1, 1
    for _ in range(3, n + 1):
        f1, f2 = f2, f1 + f2
    return f2
```

```python
def Fib4(n, memoria=None):
    if memoria is None:
        memoria = {1: 1, 2: 1}
    if n in memoria:
        return memoria[n]
    memoria[n] = Fib4(n-1, memoria) + Fib4(n-2, memoria)
    return memoria[n]
```

## Ejercicio 3) MCD

```python
def prime_factors(n):
    factors = []
    divisor = 2     # Comienza con el menor factor primo posible

    while divisor * divisor <= n: # Contina hasta que el divisor sea mayor que la raz cuadrada de 'n'
        while n % divisor == 0: # Mientras el divisor sea un factor de 'n', divide 'n' y guarda el divisor
            n = n // divisor   # Divide 'n' entre 'divisor' (quedas con un numero entero)
            factors.append(divisor)  # Aade el factor primo a la lista
        
        divisor += 1 # Incrementa el divisor para comprobar el siguiente nmero primo potencial

    # Si queda algn nmero mayor que 1, es un factor primo
    if n > 1:
        factors.append(n)  # Aade el ltimo factor primo, si lo hay
    return factors
```

```python
# Definicin de la funcin mcd1, que utiliza la descomposicin en factores primos para calcular el MCD.
def mcd1(a, b):
    # Obtenemos los factores primos de ambos nmeros
    a_factors = prime_factors(a)
    b_factors = prime_factors(b)
    
    # Encuentra los factores comunes
    common_factors = set(a_factors) & set(b_factors)
    
    # Calcula el MCD multiplicando los factores comunes elevados al menor exponente que aparecen en ambos nmeros
    mcd = 1
    for factor in common_factors:
        cantidad_factor_en_a = a_factors.count(factor)
        cantidad_factor_en_b = b_factors.count(factor)
        minima_cantidad = min(cantidad_factor_en_a, cantidad_factor_en_b)
        mcd *= factor ** minima_cantidad
    return mcd


def mcd2(a, b):
    while b != 0:
        # El nuevo 'a' es 'b' y el nuevo 'b' es el resto de la divisin de 'a' entre 'b'
        a, b = b, a % b
    # Cuando 'b' se convierte en 0, 'a' contiene el MCD
    return a
```

## Ejercicio 4) Nmeros Romanos

```python
def r2i(roman):

    roman_to_int = {'M': 1000, 'D': 500, 'C': 100, 'L': 50, 'X': 10, 'V': 5, 'I': 1}
    
    # Caso base: cadena vaca
    if not roman:
        return 0
    
    # Si la cadena tiene un solo carcter, devuelve su valor
    if len(roman) == 1:
        return roman_to_int[roman]
    
    # Caso de sustraccin: numeral menor antes de uno mayor
    if roman_to_int[roman[0]] < roman_to_int[roman[1]]:
        return r2i(roman[1:]) - roman_to_int[roman[0]]
    
    # Caso de adicin: numeral mayor o igual antes
    return roman_to_int[roman[0]] + r2i(roman[1:])

def i2r(numero):

    roman_numerals = [
        ('M', 1000), ('D', 500), ('C', 100), ('L', 50),
        ('X', 10), ('V', 5), ('I', 1)
    ]
    
    # Tuplas para casos de sustraccin
    substractive_numerals = [('C', 100), ('X', 10), ('I', 1)]

    # Caso base: nmero es 0
    if numero == 0:
        return ""
    
    # Encuentra el mayor valor romano menor o igual que 'numero'
    for numeral, value in roman_numerals:
        if value <= numero:
            return numeral + i2r(numero - value)
        
        # Verifica si puede ser escrito como sustraccin
        for substraction_numeral, substraction_value in substractive_numerals:
            # Evita la repeticin de pares de sustraccin
            if value * 0.1 <= substraction_value < value:
                substracted_value = value - substraction_value
                # Asegura un numeral de sustraccin vlido
                if 0 < substracted_value <= numero:
                    return substraction_numeral + numeral + i2r(numero - substracted_value)

```

## Ejercicio 5) Series

### a)

\begin{aligned}
& \textbf{\underline{1) Paso Base:}} \\
& \quad S_1 = \sum_{i=1}^{1} i = 1 = \frac{(1+1) \cdot 1}{2} \\
& \textbf{\underline{2) Paso Inductivo:}} \\
& \quad \textbf{a) Hiptesis Inductiva:} \\
& \quad\quad \text{Asumimos que } S_k = \sum_{i=1}^{k} i = \frac{(k+1) \cdot k}{2}  \text{ es cierto para algn } k \in \mathbb{N}.\ \\
& \quad \textbf{b) Tesis Inductiva:} \\
& \quad\quad \text{Debemos demostrar que } S_{k+1} = \sum_{i=1}^{k+1} i = \frac{(k+2) \cdot (k+1)}{2} \text{ es cierto.} \\
& \quad \textbf{c) Demostracin:} \\
& \quad\quad S_{k+1} = \sum_{i=1}^{k+1} i = \left( \sum_{i=1}^{k} i \right) + (k+1) \\
& \quad\quad\quad = S_k + (k+1) \quad \text{(usando la Hiptesis Inductiva)} \\
& \quad\quad\quad = \frac{(k+1) \cdot k}{2} + (k+1) \\
& \quad\quad\quad = \frac{k(k+1) + 2(k+1)}{2} \\
& \quad\quad\quad = \frac{(k+1)(k+2)}{2} \\
& \quad\quad\quad \text{Por lo tanto, si } S_k \text{ es verdadero, entonces } S_{k+1} \text{ tambin lo es.}
\end{aligned}

### b)

\begin{aligned}
& \textbf{\underline{1) Paso Base:}} \\
& \quad S_1 = \sum_{i=1}^{1} i^2 = 1^2 = 1 = \frac{(2\cdot 1 + 1)(1 + 1) \cdot 1}{6} \\
& \textbf{\underline{2) Paso Inductivo:}} \\
& \quad \textbf{a) Hiptesis Inductiva:} \\
& \quad\quad \text{Asumimos que } S_k = \sum_{i=1}^{k} i^2 = \frac{(2k+1)(k+1)k}{6} \text{ es cierto para algn } k \in \mathbb{N}.\ \\
& \quad \textbf{b) Tesis Inductiva:} \\
& \quad\quad \text{Debemos demostrar que } S_{k+1} = \sum_{i=1}^{k+1} i^2 = \frac{(2(k+1)+1)((k+1)+1)(k+1)}{6} \text{ es cierto.} \\
& \quad \textbf{c) Demostracin:} \\
& \quad\quad S_{k+1} = \sum_{i=1}^{k+1} i^2 = \left( \sum_{i=1}^{k} i^2 \right) + (k+1)^2 \\
& \quad\quad\quad = S_k + (k+1)^2 \quad \text{(usando la Hiptesis Inductiva)} \\
& \quad\quad\quad = \frac{(2k+1)(k+1)k}{6} + (k+1)^2 \\
& \quad\quad\quad = \frac{(2k+1)(k+1)k + 6(k+1)^2}{6} \\
& \quad\quad\quad = \frac{(2k^2+7k+6)(k+1)}{6} \\
& \quad\quad\quad = \frac{(k+2)(2k+3)(k+1)}{6} \\
& \quad\quad\quad = \frac{(2(k+1)+1)((k+1)+1)(k+1)}{6} \\
& \quad\quad\quad \text{Por lo tanto, si } S_k \text{ es verdadero, entonces } S_{k+1} \text{ tambin lo es.}
\end{aligned}

### c)

\begin{align*}
& \textbf{\underline{1) Paso Base:}} \\
& \quad S_1 = \sum_{i=1}^{1} \frac{1}{i(i+1)} = \frac{1}{1 \cdot (1+1)} = \frac{1}{2} = 1 - \frac{1}{1+1} \\
& \textbf{\underline{2) Paso Inductivo:}} \\
& \quad \textbf{a) Hiptesis Inductiva (para algn } k \in \mathbb{N}\text{):} \\
& \quad\quad \text{Asumimos que } S_k = \sum_{i=1}^{k} \frac{1}{i(i+1)} = 1 - \frac{1}{k+1} \text{ es cierto.} \\

& \quad\quad \text{Debemos demostrar que } S_{k+1} = \sum_{i=1}^{k+1} \frac{1}{i(i+1)} = 1 - \frac{1}{(k+1)+1} \text{ es cierto.} \\
& \quad \textbf{c) Demostracin:} \\
& \quad\quad S_{k+1} = \sum_{i=1}^{k+1} \frac{1}{i(i+1)} = S_k + \frac{1}{(k+1)((k+1)+1)} \\
& \quad\quad\quad = \left( 1 - \frac{1}{k+1} \right) + \frac{1}{(k+1)(k+2)} \quad \text{(usando la Hiptesis Inductiva)} \\
& \quad\quad\quad = 1 - \frac{1}{k+1} + \frac{1}{k+1} - \frac{1}{k+2} \\
& \quad\quad\quad = 1 - \frac{1}{k+2} \\
& \quad\quad\quad \text{Por lo tanto, si } S_k \text{ es cierto para algn } k \in \mathbb{N}, \text{ entonces } S_{k+1} \text{ tambin lo es.}
\end{align*}

## Ejercicio 7) Raz Cuadrada

```python
def raiz_cuadrada_recursiva(n, guess = 1.0):

    tolerancia = 10**(-12)
    if abs(n - guess**2) <= tolerancia:
        solution = guess
    else:
        solution = raiz_cuadrada_recursiva(n, (guess + n/guess)/2)
    return solution
```

```python
test_numbers = [1, 2, 9, 16, 20, 25, 100, 163, 10000]

for n in test_numbers:
    print(f"N = {n}\n"
          f"Guess:{raiz_cuadrada_recursiva(n)}\n"
          f"Real: {n**(0.5)}")
```
